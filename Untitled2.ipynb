```python
import torch
import torch.nn as nn
from transformers import AutoModel

class CapitalizationModel(nn.Module):
    def __init__(self, name):
        super(CapitalizationModel, self).__init__()
        # Load pre-trained ELECTRA model
        self.electra_model = AutoModel.from_pretrained(name)

        # Define a linear layer for binary classification (capitalize or not)
        self.linear_layer = nn.Linear(in_features=256, out_features=2)  # 256 is hidden size of electra-small

    def forward(self, x, attn_masks):
        # Get token embeddings from ELECTRA
        x = self.electra_model(x, attention_mask=attn_masks)[0]  # Shape: (batch_size, seq_len, hidden_dim)

        # Apply linear classification layer
        output = self.linear_layer(x)  # Shape: (batch_size, seq_len, 2)

        return output
```
